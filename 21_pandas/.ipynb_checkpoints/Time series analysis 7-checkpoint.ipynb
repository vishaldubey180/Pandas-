## Shifting And lagging

```python
import pandas as pd
```

```python
df = pd.read_csv('fb.csv')
```

```python
df
```

```python
df.shift(1) # data 
```

```python
df.set_index('Date', inplace=True)
```

```python
df
```

```python
df.shift(1)
```

```python
#Shift()== it shoft the data to the right hand side
```

```python
df.shift(2)
```

```python
# iffw e want to do reverse
```

```python
df.shift(-1)
```

```python
df['prev day price'] = df['Price'].shift(1)
```

```python
df
```

```python
df['1 day change'] = df['Price'] - df['prev day price']
```

```python
df
```

```python
df['5 days % return'] = df['Price']- df['Price'].shift(5)
```

```python
df
```

```python
df['5 days % return'] = (df['Price']- df['Price'].shift(5))*100/df['Price'].shift(5)
```

```python
df
```

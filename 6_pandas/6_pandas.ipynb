## Handling missing data:replace() function

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('weather_data.csv')
```

```python
df
```

```python
# we know that tempreature and windspeed cannot be -99999
```

```python
df_new  = df.replace([-99999,-8888,-756875,'etc'] , np.NaN)
```

```python
df_new
```

```python
# we can also replace the values based on the columns
```

```python
df_new = df.replace({'temperature':-99999,
                     'windspeed':-99999,
                    'event':'0'}, np.NaN)
```

```python
df_new
```

```python
# we can do mapping also that means we can replace values with other value
```

```python
df_new = df.replace({-99999:np.NaN,
                     '0':'Sunny'})
```

```python
df_new
```

```python
# not let;s create a scenario where temprature has unit also
```

```python
df3 = df.replace({32:'32f',
                     34:'34c',
                      7.0:'7mph',
                   -99999:np.NaN,
                     '0':'Sunny'})
```

```python
df3
```

```python
 # now if we don't want f , c  and mph with numbers then we can use 'regex'
```

```python
df_new = df3.replace('[A-Za-z]', '', regex=True)
```

```python
# In above code if it will ffind any character betwween a to z or A to Z it will remove that
```

```python
df_new
```

```python
# but here with f,c and mph It removed my data also
```

```python
# hence we have to do this operation on columns only
```

```python
df_new = df3.replace({'temperature':'[A-Za-z]',
                       'windspeed':'[A-Za-z]'}, '', regex=True)
```

```python
df_new
```

```python
df = pd.read_csv('weather_data.csv', parse_dates = ['day'])
```

```python
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df1 = df.fillna(0) # here it fill all nan values with 0
```

```python
df1
```

```python
# we can fill or replace values by column wise also
```

```python
df2 = df.fillna({'temperature':0,
                  'windspeed':0,
                   'event': 'no event'})
```

```python
df2
```

```python
df3 = df.fillna(method='ffill') # here if row find with 0 value than it will consider previous day value
```

```python
df3
```

```python
df4 = df.fillna(method='bfill') # here if row find with 0 value than it will consider next day value
```

```python
df4
```

```python
df5 = df.fillna(method='bfill', axis=1) # here axis is copying data horizontaly from next dat
```

```python
df5
```

```python
df6 = df.fillna(method='ffill', limit=1) # here it will copy value to only one row
```

```python
df6
```

```python
df6 = df.fillna(method='ffill', limit=2)
```

```python
df6
```

### Pandas dataframe.interpolate() function is basically used to fill NA values in the dataframe or series. But, this is a very powerful function to fill the missing values. It uses various interpolation technique to fill the missing values rather than hard-coding the value.|

#### Interpolation is the process of finding a value between two points on a line or a curve. To help us remember what it means, we should think of the first part of the word, 'inter,' as meaning 'enter,' which reminds us to look 'inside' the data we originally had. This tool, interpolation, is not only useful in statistics, but is also useful in science, business, or when there is a need to predict values that fall within two existing data points.

```python
new_df = df6.interpolate()
```

```python
new_df 
```

```python
new_df = df4.interpolate()
```

```python
new_df
```

```python
new_df = df4.interpolate(method = 'time')
```

```python
new_df
```

```python
new_df = df.interpolate(method='time')
```

```python
new_df
```

## Melt

### similar to pivot table melt is also used to transform and reshape datframe

```python
import pandas as pd
```

```python
df = pd.read_csv('weather.csv')
```

```python
df
```

```python
# data contain temperature of 3 cities on different day
```

```python
df = pd.melt(df, id_vars=['day'])
```

```python
df
```

```python
df[df['variable']=='chennai']
```

```python
df = pd.melt(df, id_vars=['day'], var_name='city', value_name='temperture')
```

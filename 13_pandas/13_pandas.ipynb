### crosstab() function
  #### The crosstab() function is used to compute a simple cross tabulation of two (or more) factors.

  ##### By default computes a frequency table of the factors unless an array of values and an aggregation function are passed.

```python
import pandas as pd
```

```python
df = pd.read_excel('survey.xls')
```

```python
df
```

```python
pd.crosstab(df.Nationality, df.Handedness)
```

```python
# so here crosstab is showing frequencey of number of variable
```

```python
pd.crosstab(df.Sex, df.Handedness)
```

```python
# if we want total also then we have to provide margin
```

```python
pd.crosstab(df.Sex, df.Handedness, margins=True)
```

```python
pd.crosstab(df.Nationality, df.Handedness, margins=True)
```

```python
# u can provide multivariable also
```

```python
pd.crosstab(df.Sex, [df.Nationality, df.Handedness], margins=True)
```

```python
pd.crosstab([df.Sex, df.Nationality], df.Handedness, margins=True)
```

```python
# if we want percentage also
```

```python
pd.crosstab(df.Sex, df.Handedness, normalize='index')
```

```python
import numpy as np
pd.crosstab(df.Sex, df.Handedness, values=df.Age, aggfunc=np.average)
```

```python
# here it is giving average age 
```

### Time Series Analysis 4

#### most common problem in data analysis is lack of uniformity in the structure of input data

#### Ex: 5th january 2017 can be represented in different ways 

```python
import pandas as pd
```

```python
dates = ['2017-01-05', 'jan 5, 2017', '01/05/2017', '2017.01.05', '2017/01/05']
```

```python
pd.to_datetime(dates)  #Here we converted our dattes in normal formate
```

### We can handel time also

```python
dates = ['2017-01-05 2:30:00 PM', 'jan 5, 2017 14:30:00', '01/05/2017', '2017.01.05', '2017/01/05']
```

```python
pd.to_datetime(dates)  #Here we converted our dattes in normal formate
```

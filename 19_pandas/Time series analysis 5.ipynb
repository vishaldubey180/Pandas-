## Let's look these things by example

```python
import pandas as pd
```

```python
df = pd.read_csv('wmt.csv')
```

```python
df
```

```python
df.set_index('Line Item', inplace=True)
```

```python
df
```

```python
## now we can transpose the table (row into column and column into row)
```

```python
df = df.T
```

```python
df
```

```python
df.index
```

```python
## here we can see that our index is object type
```

```python
# now we convert this index into period index
```

```python
df.index = pd.PeriodIndex(df.index, freq='Q-Jan')
```

```python
df
```

```python
df.index
```

```python
df['start date'] = df.index.map(lambda x: x.start_time)
```

```python
df
```

```python
df['end date'] = df.index.map(lambda x: x.end_time)
```

```python
df
```
